#### 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

x = range (1, 100)
y = range (1, 100)

plt.plot(x, y)
plt.xlabel('The x - axis')
plt.ylabel('The y - axis')
plt.title("Title for the plot")
plt.show()
```

#### 2. Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python
import matplotlib.pyplot as plt

x1 =  [10,  30, 50]
y1 =  [10,  30, 10]

x2 =  [10, 30, 50]
y2 =  [40, 10, 30]

plt.plot(x1, y1, label = "first line")
plt.plot(x2, y2, label = "second line")

plt.xlabel('The x - axis')
plt.ylabel('The y - axis')
plt.title("Title for the plot with two lines")

plt.legend()
plt.show()
```

#### 3. Write a Python program to plot two or more lines with legends, different widths and colors.



```python
import matplotlib.pyplot as plt

x1 =  [0,  30, 50]
y1 =  [10,  30, 10]

x2 =  [0, 30, 50]
y2 =  [40, 10, 30]

x3 = [0,  20, 50]
y3 = [25, 35, 0]


plt.plot(x1, y1, linewidth = 2,label = "first line")
plt.plot(x2, y2, linewidth = 7,label = "second line")
plt.plot(x3, y3, linewidth = 11,label = "third line")
plt.xlabel('The x - axis')
plt.ylabel('The y - axis')
plt.title("Title for the plot with two lines with different widths")

plt.legend()
plt.show()
```

#### 4. Write a Python program to plot two or more lines with different styles.


```python
import matplotlib.pyplot as plt

x1 =  [0,  30, 50]
y1 =  [10,  30, 10]

x2 =  [0, 30, 50]
y2 =  [40, 10, 30]

x3 = [0,  20, 50]
y3 = [25, 35, 0]


plt.plot(x1, y1, label = "first line", linestyle = 'dashdot')
plt.plot(x2, y2, label = "second line", linestyle = 'dotted')
plt.plot(x3, y3, label = "third line", linestyle = 'dashed')
plt.xlabel('The x - axis')
plt.ylabel('The y - axis')
plt.title("Title for the plot with two lines with different line styles")

plt.legend()
plt.show()
```

#### 5. Write a Python program to plot two or more lines and set the line markers.


```python
import matplotlib.pyplot as plt

x1 =  [0,  30, 50]
y1 =  [10,  30, 10]

x2 =  [0, 30, 50]
y2 =  [40, 10, 30]

x3 = [0,  20, 50]
y3 = [25, 35, 0]


plt.plot(x1, y1, label = "first line", linestyle = 'dashdot', marker = 'o')
plt.plot(x2, y2, label = "second line", linestyle = 'dotted', marker = 'v')
plt.plot(x3, y3, label = "third line", linestyle = 'dashed', marker = '<')

plt.xlabel('The x - axis')
plt.ylabel('The y - axis')
plt.title("Title for the plot with two lines with line markers")

plt.legend()
plt.show()
```

#### 6. Write a Python programming to display a bar chart of the popularity of programming Languages.
    Sample data:
    Programming languages: Java, Python, PHP, JavaScript, C#, C++
    Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7



```python
import matplotlib.pyplot as plt

Programming_Language  =  ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity =  [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x = range(len(Programming_Language))
plt.bar(x, Popularity)

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language Worldwide, Oct 2017 compared to a year ago")

plt.xticks(x, Programming_Language)
plt.show()
```

#### 7. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
    Sample data:
    Programming languages: Java, Python, PHP, JavaScript, C#, C++

#### 8. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
    Sample Data:
    Means (men) = (22, 30, 35, 35, 26)
    Means (women) = (25, 32, 30, 35, 29)


```python
import matplotlib.pyplot as plt
import numpy as np

Means_men = (22, 30, 35, 35, 26)
Means_women = (25, 32, 30, 35, 29)
num = len(Means_men)

r1,r2 = plt.subplots()
index = np.arange(num)

data1 = plt.bar(index, Means_men, 0.3, color='green',label='Men')

data2 = plt.bar(index + 0.3, Means_women, 0.3, color='red',label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + 0.3, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()

plt.show()
```

#### 9. Write a Python programming to create a pie chart of the popularity of programming Languages.



```python
import matplotlib.pyplot as plt

Programming_language  =  ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity =  [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

margins = (0.1, 0, 0, 0, 0, 0)

plt.pie(Popularity, labels=Programming_language, colors=colors, explode=margins,
autopct='%1.1f%%', shadow=True, startangle=100)

plt.axis('equal')
plt.show()
```
